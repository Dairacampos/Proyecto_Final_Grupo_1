```python
df.iloc[:, 3] = df.iloc[:, 3].astype(str).str.replace(",", ".").astype(float)
```

```python
df.iloc[:, 3] = df.iloc[:, 3].astype(float)
```

```python
def fcosto3(n):
    n1=int(n[4])
    c100derivado=(2*n1*100)+2
    return float(c100derivado)
```

```python
df["Punto 3"]=df.apply(fcosto3, axis=1)
```

```python
df
```

```python
def calificación3(n):
    n1=int(n[3])
    n2=int(n[9])
    if n1==n2:
        return 5
    else:
        return 1
```

```python
df["Calificación P3"]=df.apply(calificación3, axis=1)
```

```python
df
```

```python
def calificaciónfinal(n):
    n1=int(n[6])
    n2=int(n[8])
    n3=int(n[10])
    nfinal=(n1+n2+n3)/3
    return nfinal
```

```python
df["Calificación Final"]=df.apply(calificaciónfinal, axis=1)
```

```python
df
```

```python
df.to_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv", index=False)
```

```python
def percentiles(archivo, nombre_columna):
    try:
        df = pd.read_csv(archivo)
        if nombre_columna not in df.columns:
            return 'Error al procesar la columna'
        
        est = df[nombre_columna].describe()
        return f"Percentil 25: {est['25%']}, Percentil 50: {est['50%']}, Percentil 75: {est['75%']}"
    
    except FileNotFoundError:
        return 'Error al procesar la columna'
    except Exception as e:
        return 'Error al procesar la columna'
```

```python
percentiles("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv","CP3")
```

```python
import os
ruta = "/workspaces/Proyecto_Final_Grupo_1/pandas/Figuras"
os.makedirs(ruta, exist_ok=True)

df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")

p = range(10, 100, 10)
val = df["NF"].quantile([x/100 for x in p])

val.plot(kind="bar", color="skyblue")
plt.title("Percentiles en NF")
plt.ylabel("NF")
plt.xticks(range(len(p)), [f"{x}%" for x in p])
plt.tight_layout()

plt.savefig(os.path.join(ruta, "Percentiles.png"))
plt.close()
```
