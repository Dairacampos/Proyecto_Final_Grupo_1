```python
def c2(x):
    n=int(x[4])
    s= n*101**2+2*101+23
    s1=n*100**2+2*100+23
    sr=(s-s1)
    return float(sr)
```

```python
dn["Punto 2"]=dn.apply(c2,axis=1)
```

```python
def cal2(x):
    n=int(x[2])
    n1=int(x[7])
    if n!=n1:
        return 1
    else:
        return 5
```

```python
dn["Calificacion 2"]=dn.apply(cal2,axis=1)
```

```python
def desviacion(archivo, columna):
    try:
        ds = pd.read_csv(archivo)
        if columna not in ds.columns:
            return "Error al procesar la columna"
        return ds[columna].std()
    except FileNotFoundError:
        return "Error al procesar la columna"
    except Exception as e:
        return f"Error al procesar la columna"
```

```python
desviacion("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv","CP2")
```

```python
import os
ruta = "/workspaces/Proyecto_Final_Grupo_1/pandas/Figuras"
os.makedirs(ruta, exist_ok=True)

df = pd.read_csv("Quiz Cálculo 1- Gobierno 22 de Octubre(1-18).csv")

desviaciones = df[['CP1', 'CP2', 'CP3', 'NF']].std()

plt.figure(figsize=(8,6))
desviaciones.plot(kind='bar', color='skyblue')
plt.title('Desviaciones')
plt.ylabel('Desviación estándar')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.savefig(os.path.join(ruta, "Desviaciones.png"))
plt.close()
```
